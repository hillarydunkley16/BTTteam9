# Handling missing values

Crash Data

```python
df_crashData.head()
```

```python
nan_count_crashData = np.sum(df_crashData.isnull(), axis = 0)
nan_detected_crashData = nan_count_crashData != 0 
print(nan_detected_crashData)
```

```python
print("YEAR: ", df_crashData['YEAR'].dtype)
print("VH: ", df_crashData['VH'].dtype)
print("CYC: ", df_crashData['CYC'].dtype)
print("PED: ", df_crashData['PED'].dtype)
print("FATAL: ", df_crashData['FATAL'].dtype)
print("SERIOUS_INJURY: ", df_crashData['SERIOUS_INJURY'].dtype)
print("MINOR_INJURY: ", df_crashData['MINOR_INJURY'].dtype)
print("NO_INJURY: ", df_crashData['NO_INJURY'].dtype)
```

```python
# replace null values of float types with the mean

# compute mean for all non null YEAR, VH, CYC, PED, FATAL, SERIOUS_INJURY, MINOR_INJURY, NO_INJURY values
mean_year = df_crashData['YEAR'].mean()
print("mean value for all YEAR columns: " + str(mean_year))

mean_vh = df_crashData['VH'].mean()
print("mean value for all VH columns: " + str(mean_vh))

mean_cyc = df_crashData['CYC'].mean()
print("mean value for all CYC columns: " + str(mean_cyc))

mean_ped = df_crashData['PED'].mean()
print("mean value for all PED columns: " + str(mean_ped))

mean_fatal = df_crashData['FATAL'].mean()
print("mean value for all FATAL columns: " + str(mean_fatal))

mean_serious_injury = df_crashData['SERIOUS_INJURY'].mean()
print("mean value for all SERIOUS_INJURY columns: " + str(mean_serious_injury))

mean_minor_injury = df_crashData['MINOR_INJURY'].mean()
print("mean value for all MINOR_INJURY columns: " + str(mean_minor_injury))

mean_no_injury = df_crashData['NO_INJURY'].mean()
print("mean value for all NO_INJURY columns: " + str(mean_no_injury))
```

```python
# fill all missing values with the mean
df_crashData['YEAR'].fillna(value=mean_year, inplace=True)
df_crashData['VH'].fillna(value=mean_vh, inplace=True)
df_crashData['CYC'].fillna(value=mean_cyc, inplace=True)
df_crashData['PED'].fillna(value=mean_ped, inplace=True)
df_crashData['FATAL'].fillna(value=mean_fatal, inplace=True)
df_crashData['SERIOUS_INJURY'].fillna(value=mean_serious_injury, inplace=True)
df_crashData['MINOR_INJURY'].fillna(value=mean_minor_injury, inplace=True)
df_crashData['NO_INJURY'].fillna(value=mean_no_injury, inplace=True)
```

```python
# check that all null values have been replaced
np.sum(df_crashData.isnull(), axis=0)
```
