```python
import pandas as pd
import numpy as np
import os 
import scipy.stats as stats
```

```python
crash_data_LA = os.path.join(os.getcwd(), "data", "Crash_Data_LA_COUNTY.csv")
harsh_accleration_LA = os.path.join(os.getcwd(), "data", "Harsh_Acceleration_LA_COUNTY.csv")
harsh_braking_LA = os.path.join(os.getcwd(), "data", "Harsh_Braking_LA_COUNTY.csv")
poi_LA = os.path.join(os.getcwd(), "data", "POI_LA_COUNTY.csv")
df_crashData = pd.read_csv(crash_data_LA)
df_harshAccelerationData = pd.read_csv(harsh_accleration_LA)
df_harshBrakingData = pd.read_csv(harsh_braking_LA)
df_POIData = pd.read_csv(poi_LA)
df_crashData.head()
```
